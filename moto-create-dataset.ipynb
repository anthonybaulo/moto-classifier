## Get a list of image URLs

The more specific you are in your Google Search, the better the results and the less manual pruning you will have to do. You can also limit your results to show only photos by clicking on Tools and selecting Photos from the Type dropdown.

- Use this tool to remove unwanted images from search results: <br>

 - The tool will show the number of images that will be in the data set. 
 - Scroll down to load more images.
 - Copy url text into `.txt` file
 - Repeat for each make
    



## Create directory, upload URL files, Download images

Create `data` directory inside working directory

```python
path = Path('data')
path.mkdir(parents=True, exist_ok=True)
```

Upload the url txt files into `data`

Create dict with format: `{foldername (class): urls filename}`

```python
moto_dict = {
    'harley': 'harley_urls.txt',
    'honda': 'honda_urls.txt',
    'triumph': 'triumph_urls.txt',
    'kawasaki': 'kawasaki_urls.txt',
    'suzuki': 'suzuki_urls.txt',
    'yamaha': 'yamaha_urls.txt'
}
```

## Download the images

Note: If you have problems with download, try with `max_workers=0` to see exceptions:

`download_images(path/file, dest, max_pics=20, max_workers=0)`

```python
for folder, file in moto_dict.items():
    dest = path/folder
    dest.mkdir(parents=True, exist_ok=True)
    download_images(path/file, dest, max_pics=225)
```

### Remove any images that can't be opened:

```python
for c in moto_dict:
    print(c)
    verify_images(path/c, delete=True, max_size=500)
```

## View data

```python
np.random.seed(42) 

# valid_pct=0, and no transforms -  so that we can find duplicates (see below)
data = ImageDataBunch.from_folder(path, train=".", valid_pct=0., 
                                  size=224, num_workers=4)
```

```python
data.classes, data.c, len(data.train_ds), len(data.valid_ds)
```

```python
data.show_batch(rows=3, figsize=(7,8))
```

### Delete duplicates 

Create a learner to use with `DatasetFormatter()` and `ImageCleaner`

```python
learn = cnn_learner(data, models.resnet34)
```

```python
ds, idxs = DatasetFormatter().from_similars(learn, ds_type=DatasetType.Train)
```

```python
ImageCleaner(ds, idxs, path, duplicates=True)
```
